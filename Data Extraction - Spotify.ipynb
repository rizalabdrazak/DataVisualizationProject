Album Data

```python
album_name=[]
album_id=[]
track_in_album=[]
available_market=[]
album_artist_name=[]
```

```python
tes=sp.search(q='year:2018',type='album',limit=10)
```

```python
tes['albums']['items'][0]['artists'][0]['name']
```

```python
for i in range (0,1000,50):
    album_results=sp.search(q='year:2018',type='album',limit=50,offset=i)
    for i,t in enumerate(album_results['albums']['items']):
        album_name.append(t['name'])
        album_id.append(t['id'])
        track_in_album.append(t['total_tracks'])
        album_artist_name.append(t['artists'][0]['name'])
        available_market.append(t['available_markets'])
        
```

```python
print(len(album_name))
print(len(album_id))
print(len(track_in_album))
print(len(available_market))
print(len(album_artist_name))
```

```python
album_df=pd.DataFrame({
                        'album_name':album_name,
                        'album_id':album_id,
                        'track_in_album':track_in_album,
                        'available_market':available_market,
                        'album_artist_name':album_artist_name
                     })
```

```python
album_df.head()
```

```python
album_df.to_csv('album_df',sep='\t')
```

```python
artist_name=[]
track_name=[]
track_id=[]
popularity=[]
for i in range(0,10000,50):
    track_results = sp.search(q='year:2018', type='track', limit=50,offset=i)
    for i, t in enumerate(track_results['tracks']['items']):
        artist_name.append(t['artists'][0]['name'])
        track_name.append(t['name'])
        track_id.append(t['id'])
        popularity.append(t['popularity'])
```
